## schedule 모듈 공부
https://schedule.readthedocs.io/en/stable/index.html

```python
import schedule
import time
import datetime
```

```python
def job():
    print('잘 돌아가는 중입니다 ...')


# 10초에 한번씩 실행
schedule.every(10).seconds.do(job)
# 10분에 한번씩 실행
schedule.every(10).minutes.do(job)
# 매 시간 실행
schedule.every().hour.do(job)
# 매일 10:30 에 실행
schedule.every().day.at("10:30").do(job)
# 매주 월요일 실행
schedule.every().monday.do(job)
# 매주 수요일 13:15 에 실행
schedule.every().wednesday.at("13:15").do(job)
```

새롭게 바뀌는 것이 아니라 추가가 계속 된다.

```python
count = 0
while True:
    schedule.run_pending()
    count += 1
    time.sleep(1)
    print(count)
```

```python
time_list = ['17:15', '17:16','17:17']
# 모든 스케쥴 초기화
schedule.clear()
for i in range(len(time_list)):
    schedule.every().day.at(time_list[i]).do(job)
```

```python
while True:
    schedule.run_pending()
    time.sleep(1)
    print(datetime.datetime.now())
```
